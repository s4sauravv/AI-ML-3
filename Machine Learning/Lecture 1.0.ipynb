```python
#create x=feature and y=label

x=data[['TV']]
y=data.sales

#follow the usual sklearn patter : import , instantiate ,fit


from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x, y)
```

```python
#Print intercept and coefficients

print(lm.intercept_)
print(lm.coef_)
```
